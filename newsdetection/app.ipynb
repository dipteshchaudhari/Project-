1 : convert all the text into lower case
2 : remove the text which were the enclosed in the []
3 : this regular expression for replace all the non werb character like (digits,underscores) with a space 
4 : this will remove the url's which would be stating with http or https  or domain start with www. 
5 : using this we ll remove the html tags 
6 : this will remove the punctuation marks by asccaping and matching all characters in the strings.punctuation
7 : remove the new line characters 
8 : removing words which contains digits , it can be "132sdf"

```python
def clean_text(text):
    text = text.lower()         
    text = re.sub('\[.*?\]',"",text)
    text = re.sub("\\w"," ",text)
    text = re.sub("https?:://\S+|www\.\S+" "",text)
    text = re.sub("<.*?>=+","",text)
    text = re.sub("[%s]" % re.escape(string.punctuation),"",text)
    text = re.sub("\n","",text)
    text = re.sub("\w*\d\w*","",text)
    return text
```

```python

def clean_text(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)  # remove things in [ ]
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # remove URLs
    text = re.sub(r'<.*?>', '', text)  # remove HTML tags
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)  # remove punctuation
    text = re.sub(r'\n', ' ', text)  # remove newlines
    text = re.sub(r'\w*\d\w*', '', text)  # remove words with numbers
    text = re.sub(r'\W+', ' ', text)  # replace non-word chars with space
    return text.strip()
```

🔎 What is Training and Testing in Machine Learning?

Training Data

This is the part of the dataset we use to teach the model.

The model looks at the input (text) and the output (class) to learn patterns.

Example:

Input: "Trump wins election" → Output: 1 (True news)

Input: "Aliens attacked the White House" → Output: 0 (Fake news)

By seeing many such examples, the model learns what fake and true news look like.

Testing Data

This is the part of the dataset that the model has never seen before.

We use it to check how well the model performs on new, unseen news.

Example: After training, we give it a new text "Government launches new scheme" → The model predicts if it’s Fake or True.

If the model gets it right most of the time, we know it has learned wel

in below code , We are spitting four parts :

******xtrain (Training Features)*******

These are the input news texts (from "text") used for training the model.
Example: "Government announces new policy" → model will read this.

*******ytrain (Training Labels)*******

These are the correct answers (from "class") for the training inputs.

Example: 1 (True news), 0 (Fake news).

The model uses these labels to learn the relationship between text and class.

*******xtest (Testing Features)*******

These are the input news texts reserved for testing (25% of the data).

The model has never seen these before.

********ytest (Testing Labels)********

These are the correct answers for the testing inputs.

We don’t show these to the model during training.

After training, we compare the model’s predictions with ytest to calculate accuracy, precision, recall, etc.



"xtrain and ytrain are like the practice material for the model: it learns patterns from them.

xtest and ytest are like the final exam: we test the model on new data (xtest) and check if its answers match the real answers (ytest)."

```python
x = data["text"]
y = data["class"]

xtrain, xtest , ytrain, ytest = train_test_split(x,y,test_size=0.25,random_state=42) 
```

using {TF- IDF factorizer}for converting textual data to numerical data ,
 here we create a tfidfvectorizer object


```python
# Replace NaN with empty string
xtrain = xtrain.fillna("")
xtest = xtest.fillna("")
```

```python
vectorizer = TfidfVectorizer()
xv_train = vectorizer.fit_transform(xtrain)
xv_test = vectorizer.transform(xtest)
```

```python
data.sample(10)
```

```python
lr = LogisticRegression()
lr.fit(xv_train,ytrain)
```

```python
prediction = lr.predict(xv_test)
lr.score(xv_test,ytest)
```

```python
print(classification_report(ytest,prediction))
```

```python
joblib.dump(vectorizer,"vectorizer.jb")
joblib.dump(lr,"lr_model.jb")
```
